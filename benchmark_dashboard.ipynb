```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load benchmark results

```python
df = pd.read_csv("benchmark_results.csv")
```

# Basic info

```python
print("✅ Loaded results:", df.shape)
display(df.head())
```

# Plot: Activation vs R² Score

```python

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='activation', y='r2')
plt.title("Activation Function vs R² Score")
plt.grid()
plt.show()
```

# Plot: Layers vs RMSE

```python

plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='max_layers', y='rmse', hue='activation', marker='o')
plt.title("Layers vs RMSE per Activation")
plt.grid()
plt.show()
```

# Plot: Quantized vs Non-Quantized R²

```python

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='quantized', y='r2')
plt.title("Effect of Quantization on R²")
plt.grid()
plt.show()
```

# Plot: Training Time vs R²

```python

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='training_time', y='r2', hue='activation')
plt.title("Training Time vs R²")
plt.grid()
plt.show()
```

# Plot: Model Size vs MAE

```python

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='model_size_MB', y='mae', hue='quantized')
plt.title("Model Size vs MAE")
plt.grid()
plt.show()
```
